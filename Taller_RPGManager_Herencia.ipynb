```python
from abc import ABC, abstractmethod


class Personaje(ABC):

    def __init__(self, nombre: str, nivel: int, vida_max: int):
        self._nombre          = nombre
        self._nivel           = nivel
        self._puntos_vida_max = vida_max
        self._puntos_vida     = vida_max

    def recibir_dano(self, dano: int) -> None:
        self._puntos_vida = max(0, self._puntos_vida - dano)
        estado = "— ¡Ha sido derrotado!" if self._puntos_vida == 0 else ""
        print(f"{self._nombre} recibe {dano} de daño. "
              f"HP: {self._puntos_vida}/{self._puntos_vida_max} {estado}".strip())

    def esta_vivo(self) -> bool:
        return self._puntos_vida > 0

    @abstractmethod
    def atacar(self, objetivo: "Personaje") -> None: ...

    @abstractmethod
    def get_tipo_personaje(self) -> str: ...

    def __str__(self) -> str:
        return (f"[{self.get_tipo_personaje()}] {self._nombre} "
                f"Nv.{self._nivel} | HP: {self._puntos_vida}/{self._puntos_vida_max}")

    @property
    def nombre(self):          return self._nombre
    @property
    def nivel(self):           return self._nivel
    @property
    def puntos_vida(self):     return self._puntos_vida
    @property
    def puntos_vida_max(self): return self._puntos_vida_max
```

```python
class Guerrero(Personaje, Habilidoso, Equipable):

    def __init__(self, nombre: str, nivel: int):
        super().__init__(nombre, nivel, 100 + nivel * 10)
        self._fuerza        = 15 + nivel * 3
        self._defensa       = 10 + nivel * 2
        self._item_equipado  = "Sin equipo"
        self._costo_habilidad = 30

    # ── Personaje abstractos ──────────────────────────────────────────────────

    def atacar(self, objetivo: Personaje) -> None:
        print(f"{self._nombre} golpea a {objetivo.nombre} con {self._fuerza} de daño!")
        objetivo.recibir_dano(self._fuerza)

    def get_tipo_personaje(self) -> str:
        return "Guerrero"

    # ── Habilidoso ────────────────────────────────────────────────────────────

    def usar_habilidad_especial(self, objetivo: Personaje) -> None:
        print(f"{self._nombre} usa Golpe Devastador sobre {objetivo.nombre} causando 50 de daño!")
        objetivo.recibir_dano(50)

    def get_costo_habilidad(self) -> int:
        return self._costo_habilidad

    def get_nombre_habilidad(self) -> str:
        return "Golpe Devastador"

    # ── Equipable ─────────────────────────────────────────────────────────────

    def equipar_item(self, item: str) -> None:
        self._item_equipado = item
        print(f"{self._nombre} equipa: {self._item_equipado}")

    def get_item_equipado(self) -> str:
        return self._item_equipado

    # ── Propio ────────────────────────────────────────────────────────────────

    def usar_escudo(self) -> None:
        print(f"{self._nombre} bloquea con su escudo. Defensa: {self._defensa}")

    @property
    def fuerza(self):        return self._fuerza
    @property
    def defensa(self):       return self._defensa
    @property
    def item_equipado(self): return self._item_equipado
```

```python
class Mago(Personaje, Habilidoso, Sanador):

    def __init__(self, nombre: str, nivel: int):
        super().__init__(nombre, nivel, 60 + nivel * 5)
        self._mana_max = 80 + nivel * 10
        self._mana     = self._mana_max

    # ── Personaje abstractos ──────────────────────────────────────────────────

    def atacar(self, objetivo: Personaje) -> None:
        if self._mana >= 20:
            dano = 25 + self._nivel * 5
            print(f"{self._nombre} lanza un hechizo sobre {objetivo.nombre} causando {dano} de daño!")
            objetivo.recibir_dano(dano)
            self._mana -= 20
            print(f"{self._nombre} | Mana: {self._mana}/{self._mana_max}")
        else:
            print(f"{self._nombre} no puede lanzar el hechizo — mana insuficiente!")

    def get_tipo_personaje(self) -> str:
        return "Mago"

    # ── Habilidoso ────────────────────────────────────────────────────────────

    def usar_habilidad_especial(self, objetivo: Personaje) -> None:
        if self._mana >= 20:
            print(f"{self._nombre} lanza Bola de Fuego sobre {objetivo.nombre} causando 40 de daño!")
            objetivo.recibir_dano(40)
            self._mana -= 20
            print(f"{self._nombre} | Mana: {self._mana}/{self._mana_max}")
        else:
            print(f"{self._nombre} no tiene mana suficiente para Bola de Fuego!")

    def get_costo_habilidad(self) -> int:
        return 20

    def get_nombre_habilidad(self) -> str:
        return "Bola de Fuego"

    # ── Sanador ───────────────────────────────────────────────────────────────

    def sanar(self, objetivo: Personaje) -> None:
        vida_nueva = min(objetivo.puntos_vida_max, objetivo.puntos_vida + 25)
        objetivo._puntos_vida = vida_nueva
        print(f"{self._nombre} sana a {objetivo.nombre} — HP: {vida_nueva}/{objetivo.puntos_vida_max}")

    def get_potencia_sanacion(self) -> int:
        return 25

    # ── Propio ────────────────────────────────────────────────────────────────

    def recuperar_mana(self, cantidad: int) -> None:
        self._mana = min(self._mana_max, self._mana + cantidad)
        print(f"{self._nombre} recupera mana. Mana: {self._mana}/{self._mana_max}")

    @property
    def mana(self):     return self._mana
    @property
    def mana_max(self): return self._mana_max
```

```python
class Arquero(Personaje, Equipable):

    def __init__(self, nombre: str, nivel: int):
        super().__init__(nombre, nivel, 75 + nivel * 7)
        self._flechas      = 10 + nivel * 2
        self._alcance      = 30
        self._item_equipado = "Arco basico"

    # ── Personaje abstractos ──────────────────────────────────────────────────

    def atacar(self, objetivo: Personaje) -> None:
        if self._flechas > 0:
            dano = 12 + self._nivel * 4
            bonus = ""
            if self._item_equipado != "Arco basico":
                dano += 5
                bonus = " [+5 bonus]"
            print(f"{self._nombre} dispara una flecha a {objetivo.nombre} causando {dano} de daño!{bonus}")
            objetivo.recibir_dano(dano)
            self._flechas -= 1
            print(f"{self._nombre} | Flechas restantes: {self._flechas}")
        else:
            print(f"{self._nombre} no puede disparar — no le quedan flechas!")

    def get_tipo_personaje(self) -> str:
        return "Arquero"

    # ── Equipable ─────────────────────────────────────────────────────────────

    def equipar_item(self, item: str) -> None:
        self._item_equipado = item
        print(f"{self._nombre} equipa: {self._item_equipado}")

    def get_item_equipado(self) -> str:
        return self._item_equipado

    # ── Propio ────────────────────────────────────────────────────────────────

    def recargar_flechas(self, cantidad: int) -> None:
        self._flechas += cantidad
        print(f"{self._nombre} recarga {cantidad} flechas. Total: {self._flechas}")

    @property
    def flechas(self):       return self._flechas
    @property
    def alcance(self):       return self._alcance
    @property
    def item_equipado(self): return self._item_equipado
```

```python
class Habilidoso(ABC):

    @abstractmethod
    def usar_habilidad_especial(self, objetivo: "Personaje") -> None: ...

    @abstractmethod
    def get_costo_habilidad(self) -> int: ...

    @abstractmethod
    def get_nombre_habilidad(self) -> str: ...


class Equipable(ABC):

    @abstractmethod
    def equipar_item(self, item: str) -> None: ...

    @abstractmethod
    def get_item_equipado(self) -> str: ...


class Sanador(ABC):

    @abstractmethod
    def sanar(self, objetivo: "Personaje") -> None: ...

    @abstractmethod
    def get_potencia_sanacion(self) -> int: ...
```
