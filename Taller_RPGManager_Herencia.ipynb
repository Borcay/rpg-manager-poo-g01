```python
from abc import ABC, abstractmethod


class Personaje(ABC):

    def __init__(self, nombre: str, nivel: int, vida_max: int):
        self._nombre          = nombre
        self._nivel           = nivel
        self._puntos_vida_max = vida_max
        self._puntos_vida     = vida_max

    def recibir_dano(self, dano: int) -> None:
        self._puntos_vida = max(0, self._puntos_vida - dano)
        estado = "— ¡Ha sido derrotado!" if self._puntos_vida == 0 else ""
        print(f"{self._nombre} recibe {dano} de daño. "
              f"HP: {self._puntos_vida}/{self._puntos_vida_max} {estado}".strip())

    def esta_vivo(self) -> bool:
        return self._puntos_vida > 0

    @abstractmethod
    def atacar(self, objetivo: "Personaje") -> None: ...

    @abstractmethod
    def get_tipo_personaje(self) -> str: ...

    def __str__(self) -> str:
        return (f"[{self.get_tipo_personaje()}] {self._nombre} "
                f"Nv.{self._nivel} | HP: {self._puntos_vida}/{self._puntos_vida_max}")

    @property
    def nombre(self):          return self._nombre
    @property
    def nivel(self):           return self._nivel
    @property
    def puntos_vida(self):     return self._puntos_vida
    @property
    def puntos_vida_max(self): return self._puntos_vida_max
```
